```python
data = {
    "name":  ["Alice", "Bob", "Clara", "David", "Eva"],
    "age":   [21, 25, 22, 19, 23],
    "score": [85, 90, 91, 76, 60],
    "city":  ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai"]
}
```

```python
import pandas as pd

df = pd.DataFrame(data)

print(df)
```

```python
df.shape
```

```python
df.info()
```

```python
# Selecting columns

print(df["name"])
print(df[["name","age"]])
```

```python
# Selecting rows

print(df.iloc[0])
print(df.iloc[0:3])
```

```python
# Filtering rows

print(df[df["score"]>80])

print(df[df["city"]=="Mumbai"])

print(df[(df["age"]>22) & (df["score"]>70)])
```

```python
# Sorting

df.sort_values("age", ascending=False)
```

```python
# Adding New column with Grades

df["grade"] = df["score"].apply(lambda x: "A" if x >= 90 else "B" if x>=80 else "C" if x>=70 else "F")
```
