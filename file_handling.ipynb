```python
# Cleaning the rew data without Numpy & Pandas

def load_students(students):
    try:
        with open(students, "r") as f:
            lines = f.readlines()[1:]
    except FileNotFoundError:
        print("File not found")
        return []
    
    Clean_data = []
    for line in lines:
        parts = line.strip().split(",")

        # print(f"parts={parts}, len={len(parts)}") 

        # if len(parts) != 3:
        #     continue

        name, age, score = parts[0], parts[1], parts[2]

        if name=="":
            print(f"Skipped row {line.strip()} -> missing name")
            continue
        try:
            age=int(age)
        except:
            print(f"Skipped row {line.strip()} -> invalied age")
            continue
        try:
            score=int(score)
        except:
            print(f"Skipped row {line.strip()} -> invalied Score")
            continue

        if score not in range(0, 100):
            print(f"Skipped row {line.strip()} -> Score not in limit")
            continue

        Clean_data.append({"name":name, "age":age, "score":score})


    print("\nClean data:")
    print(Clean_data)
    return Clean_data

load_students("students.csv")
```
