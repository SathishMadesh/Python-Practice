```python
import pandas as pd

data = {
    "name":   ["Alice", "Bob", "Clara", "David", "Eva", "Frank"],
    "age":    [21, 25, 22, 19, 23, None],
    "score":  [85, 90, None, 76, 60, 88],
    "city":   ["Delhi", "Mumbai", "Delhi", "Chennai", "Mumbai", "Delhi"]
}

df = pd.DataFrame(data)
```

```python
# Print the shape and basic info of the DataFrame

print(df.shape)
```

```python
# Print all students from "Delhi"

print(df[df["city"]=="Delhi"])
```

```python
# Print the mean score of each city using groupby

df.groupby("city")["score"].mean()
```

```python
# Fill missing age with the mean age, and missing score with the mean score

df["age"]=df["age"].fillna(df["age"].mean())

df["score"]=df["score"].fillna(df["score"].mean())
```

```python
# Add a new column "grade" using the same logic as before — A/B/C based on score

df["grade"] = df["score"].apply(lambda x: "A" if x>=90 else "B" if x>=80 else "C" if x>=70 else "F")

df.head()
```

```python
# Sort the DataFrame by score in descending order and print it

print(df.sort_values("score", ascending = False))
```
