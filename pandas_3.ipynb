```python
import pandas as pd

orders = pd.DataFrame({
    "order_id":    [1, 2, 3, 4, 5, 6],
    "customer":    ["Alice", "Bob", "Alice", "Clara", "Bob", "Alice"],
    "product_id":  [101, 102, 103, 101, 103, 102],
    "amount":      [250, 180, 320, 150, 210, 400]
})

products = pd.DataFrame({
    "product_id":   [101, 102, 103],
    "product_name": ["Laptop Stand", "Mouse", "Keyboard"]
})
```

```python
df = pd.DataFrame(pd.merge(orders, products, on="product_id", how="inner"))

print(df)
```

```python
print(df.groupby("customer")["amount"].sum())
```

```python
print(df["product_name"].value_counts())
```

```python
df["tag"] = df["amount"].apply(lambda x: "big" if x>=300 else "small")

print(df)
```

```python
df["customer_upper"] = df["customer"].str.upper()
print(df)
```

```python
print(df.sort_values("amount", ascending=False)) 
```
