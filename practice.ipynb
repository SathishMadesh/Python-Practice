```python
## Count Frequency of Characters in a String

s = "datascience"
freq = {}

def count_char(word):
    for char in word:
        if char in freq:
            freq[char] = freq[char] + 1
        else:
            freq[char] = 1

    for key, value in freq.items():
        print(key, ":", value)

count_char(s)
```

```python
## List

nums = [2, 4, 6, 8]

def sum_of_list(list):
    total = 0
    for num in list:
        total = total + num
    return total

sum_of_list(nums)
```

```python
## Find the Largest Element in a List

nums = [-3, -1, -9, -4, -7]

def larg_in_list(lst):
    max = lst[0]
    for num in lst:
        if num >= max:
            max = num
    return max

larg_in_list(nums)
```

```python
## Reverse a List (Without using reverse() or slicing)

nums = [1, 2, 3, 4]

def reverse_a_list(lst):
    new_list = []
    for num in nums:
        new_list = [num] + new_list
    return new_list

reverse_a_list(nums)
```

```python
## how many times 2 appears

nums = [1,2,3,2,4,2,5]
target = 2

def count_of_number(lst, target):
    count = 0
    for num in lst:
        if num == target:
            count += 1
    return count

count_of_number(nums, target)
```
