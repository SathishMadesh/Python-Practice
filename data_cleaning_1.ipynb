```python
import numpy as np
```

```python
scores = np.array([45, 82, 91, 60, 73, 55, 88, 76, 49, 95])

# Using only NumPy (no loops, no list comprehensions):

# Print the mean, max, min and standard deviation
# Print all scores above 70
# Print how many students scored above 70
# Replace all scores below 60 with 0 (they failed)
# Print the normalized scores — formula: (score - mean) / std

mean_score = np.mean(scores)
max_score = np.max(scores)
min_score = np.min(scores)
print(mean_score)
print(max_score)
print(min_score)

scores_abv_70 = scores[scores > 70]

std_scr_abv_70 = np.sum(scores > 70)

scores[scores < 60] = 0

## Normalized score

normalized_scores = (scores-np.mean(scores))/np.std(scores)

print(scores_abv_70)
print(std_scr_abv_70)
print (scores)
print(normalized_scores)
```
